```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

## A. Estructura de una Grafica


```python
# A1) Figura con plt.figure()
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig = plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='royalblue')
plt.title('Figura creada con plt.figure()')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.annotate('Pico', xy=(1.57, 1), xytext=(3.2, 1.2), arrowprops=dict(arrowstyle='->'))
plt.grid(alpha=0.3)
plt.show()
```

```python
# A2) Figura con plt.subplots()
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.cos(x), label='cos(x)', color='tomato')
ax.set_title('Figura creada con plt.subplots()')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
ax.annotate('Valle', xy=(3.14, -1), xytext=(4.4, -1.2), arrowprops=dict(arrowstyle='->'))
ax.grid(alpha=0.3)
plt.show()
```

```python
# A3) Ejes con add_subplot()
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, np.sin(x), color='purple', label='sin(x)')
ax1.set_title('add_subplot - Eje 1')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.legend()

ax2.plot(x, np.cos(x), color='green', label='cos(x)')
ax2.set_title('add_subplot - Eje 2')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.legend()

fig.suptitle('Figura con add_subplot()')
fig.tight_layout()
plt.show()
```

## B. Interfaz Pyplot vs API OO


```python
# B1) Pyplot
x2 = np.linspace(0, 5, 100)
plt.figure(figsize=(7, 4))
plt.plot(x2, x2**2, color='darkorange', label='x^2')
plt.title('Pyplot basico')
plt.xlabel('x')
plt.ylabel('x^2')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# B2) OO
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x2, x2**2, color='teal', label='x^2')
ax.set_title('API OO basica')
ax.set_xlabel('x')
ax.set_ylabel('x^2')
ax.legend()
ax.grid(alpha=0.3)
plt.show()
```

## C. Graficas Univariantes


```python
datos_uni = np.random.normal(loc=50, scale=12, size=500)
x3 = np.linspace(0, 2 * np.pi, 300)
y_sen = np.sin(x3)
y_cos = np.cos(x3)
```

```python
# C1) Histograma basico
plt.figure(figsize=(7, 4))
plt.hist(datos_uni, bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma basico')
plt.show()
```

```python
# C2) Histograma acumulativo
plt.figure(figsize=(7, 4))
plt.hist(datos_uni, bins=20, cumulative=True, color='salmon', edgecolor='black')
plt.title('Histograma acumulativo')
plt.show()
```

```python
# C3) Histograma normalizado
plt.figure(figsize=(7, 4))
plt.hist(datos_uni, bins=20, density=True, color='mediumseagreen', edgecolor='black')
plt.title('Histograma normalizado')
plt.show()
```

```python
# C4) Boxplot basico
plt.figure(figsize=(6, 4))
plt.boxplot(datos_uni)
plt.title('Boxplot basico')
plt.show()
```

```python
# C5) Boxplot personalizado
plt.figure(figsize=(6, 4))
bp = plt.boxplot(datos_uni, patch_artist=True)
for box in bp['boxes']:
    box.set(facecolor='lightyellow', edgecolor='darkred', linewidth=2)
plt.title('Boxplot personalizado')
plt.show()
```

```python
# C6) Violin Plot basico
plt.figure(figsize=(6, 4))
plt.violinplot(datos_uni, showmeans=True)
plt.title('Violin Plot basico')
plt.show()
```

```python
# C7) Violin Plot personalizado
plt.figure(figsize=(6, 4))
vp = plt.violinplot(datos_uni, showmeans=True, showmedians=True)
for body in vp['bodies']:
    body.set_facecolor('lightblue')
    body.set_edgecolor('navy')
plt.title('Violin Plot personalizado')
plt.show()
```

```python
# C8) Linea simple seno
plt.figure(figsize=(8, 4))
plt.plot(x3, y_sen)
plt.title('Linea simple: seno')
plt.show()
```

```python
# C9) Linea personalizada
plt.figure(figsize=(8, 4))
plt.plot(x3, y_sen, color='crimson', linestyle='--', marker='o', markevery=25)
plt.title('Linea personalizada')
plt.show()
```

```python
# C10) Seno vs coseno
plt.figure(figsize=(8, 4))
plt.plot(x3, y_sen, label='sin(x)')
plt.plot(x3, y_cos, label='cos(x)')
plt.title('Comparativa seno vs coseno')
plt.legend()
plt.show()
```

## D. Graficas Bivariantes


```python
x_bi = np.linspace(0, 10, 120)
y_bi = 2.5 * x_bi + np.random.normal(0, 3, size=x_bi.size)
tam = np.random.randint(20, 300, size=x_bi.size)
color_val = np.random.rand(x_bi.size)
```

```python
# D1) Scatter simple
plt.figure(figsize=(7, 4))
plt.scatter(x_bi, y_bi, alpha=0.7)
plt.title('Scatter simple')
plt.show()
```

```python
# D2) Scatter con tamano y color
plt.figure(figsize=(7, 4))
plt.scatter(x_bi, y_bi, s=tam, c=color_val, alpha=0.7)
plt.title('Scatter s y c')
plt.show()
```

```python
# D3) Scatter con mapa de color
plt.figure(figsize=(7, 4))
sc = plt.scatter(x_bi, y_bi, c=color_val, cmap='viridis', s=80)
plt.colorbar(sc)
plt.title('Scatter con cmap')
plt.show()
```

```python
# D4) Scatter con regresion
coef = np.polyfit(x_bi, y_bi, 1)
y_reg = np.polyval(coef, x_bi)
plt.figure(figsize=(7, 4))
plt.scatter(x_bi, y_bi, alpha=0.6)
plt.plot(x_bi, y_reg, color='red')
plt.title('Scatter con regresion lineal')
plt.show()
```

```python
# D5) Linea comparativa de dos variables
x_cmp = np.arange(1, 13)
var1 = np.random.randint(30, 80, size=12)
var2 = np.random.randint(20, 70, size=12)
plt.figure(figsize=(8, 4))
plt.plot(x_cmp, var1, marker='o', label='A')
plt.plot(x_cmp, var2, marker='s', label='B')
plt.legend()
plt.title('Comparativa dos variables')
plt.show()
```

```python
# D6) Regresion cuadratica
x_q = np.linspace(-4, 4, 120)
y_q = 0.8 * x_q**2 + 1.5 * x_q + 5 + np.random.normal(0, 2, size=x_q.size)
coef2 = np.polyfit(x_q, y_q, 2)
y_fit2 = np.polyval(coef2, x_q)
plt.figure(figsize=(7, 4))
plt.scatter(x_q, y_q, s=25, alpha=0.6)
plt.plot(x_q, y_fit2, color='darkgreen', linewidth=2.5)
plt.title('Regresion cuadratica')
plt.show()
```

```python
# D7) Heatmap basico
mat = np.random.rand(8, 8)
plt.figure(figsize=(5, 4))
plt.imshow(mat)
plt.colorbar()
plt.title('Heatmap basico')
plt.show()
```

```python
# D8) Heatmap personalizado
plt.figure(figsize=(6, 4))
plt.imshow(mat, cmap='coolwarm', interpolation='nearest')
plt.xticks(range(mat.shape[1]), [f'C{i+1}' for i in range(mat.shape[1])], rotation=45)
plt.yticks(range(mat.shape[0]), [f'F{i+1}' for i in range(mat.shape[0])])
plt.colorbar()
plt.title('Heatmap personalizado')
plt.tight_layout()
plt.show()
```

```python
# D9) Heatmap correlacion
v1 = np.random.normal(0, 1, 200)
v2 = 0.7 * v1 + np.random.normal(0, 0.7, 200)
v3 = np.random.normal(0, 1, 200)
v4 = -0.5 * v1 + 0.3 * v3 + np.random.normal(0, 0.6, 200)
mat_corr = np.corrcoef(np.vstack([v1, v2, v3, v4]))
plt.figure(figsize=(6, 5))
plt.imshow(mat_corr, cmap='RdBu_r', vmin=-1, vmax=1)
plt.xticks(range(4), ['v1', 'v2', 'v3', 'v4'])
plt.yticks(range(4), ['v1', 'v2', 'v3', 'v4'])
plt.colorbar()
plt.title('Heatmap de correlacion')
plt.show()
```

```python
# D10) Heatmap con anotaciones
plt.figure(figsize=(6, 5))
plt.imshow(mat_corr, cmap='YlGnBu', vmin=-1, vmax=1)
plt.xticks(range(4), ['v1', 'v2', 'v3', 'v4'])
plt.yticks(range(4), ['v1', 'v2', 'v3', 'v4'])
for i in range(mat_corr.shape[0]):
    for j in range(mat_corr.shape[1]):
        plt.text(j, i, f'{mat_corr[i, j]:.2f}', ha='center', va='center')
plt.colorbar()
plt.title('Heatmap anotado')
plt.show()
```

## E. Graficas Categoricas


```python
categorias = ['A', 'B', 'C', 'D']
serie1 = np.array([15, 20, 14, 10])
serie2 = np.array([8, 12, 9, 7])
serie3 = np.array([5, 7, 6, 4])
```

```python
# E1) Barras apiladas
plt.figure(figsize=(7, 4))
plt.bar(categorias, serie1, label='Serie 1')
plt.bar(categorias, serie2, bottom=serie1, label='Serie 2')
plt.bar(categorias, serie3, bottom=serie1 + serie2, label='Serie 3')
plt.title('Barras apiladas')
plt.legend()
plt.show()
```

```python
# E2) Barras agrupadas
x_cat = np.arange(len(categorias))
w = 0.25
plt.figure(figsize=(8, 4))
plt.bar(x_cat - w, serie1, width=w, label='Serie 1')
plt.bar(x_cat, serie2, width=w, label='Serie 2')
plt.bar(x_cat + w, serie3, width=w, label='Serie 3')
plt.xticks(x_cat, categorias)
plt.title('Barras agrupadas')
plt.legend()
plt.show()
```

```python
# E3) Barras apiladas personalizadas
plt.figure(figsize=(7, 4))
plt.bar(categorias, serie1, color='lightcoral', hatch='//', edgecolor='black', label='Serie 1')
plt.bar(categorias, serie2, bottom=serie1, color='lightskyblue', hatch='..', edgecolor='black', label='Serie 2')
plt.bar(categorias, serie3, bottom=serie1 + serie2, color='lightgreen', hatch='xx', edgecolor='black', label='Serie 3')
plt.title('Barras personalizadas')
plt.legend()
plt.show()
```

```python
# E4) Barras con etiquetas numericas
fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(categorias, serie1, label='Serie 1')
ax.bar(categorias, serie2, bottom=serie1, label='Serie 2')
ax.bar(categorias, serie3, bottom=serie1 + serie2, label='Serie 3')
for i in range(len(categorias)):
    ax.text(i, serie1[i] / 2, str(serie1[i]), ha='center', va='center', color='white')
    ax.text(i, serie1[i] + serie2[i] / 2, str(serie2[i]), ha='center', va='center')
    ax.text(i, serie1[i] + serie2[i] + serie3[i] / 2, str(serie3[i]), ha='center', va='center')
ax.set_title('Barras con etiquetas')
ax.legend()
plt.show()
```
